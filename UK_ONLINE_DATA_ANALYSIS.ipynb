```python
import pandas as pd
filepath ="Uk_Online_Retail_Data.xlsx"


df_2009 =pd.read_excel(filepath,sheet_name='Year 2009-2010')
df_2010= pd.read_excel(filepath,sheet_name='Year 2010-2011')


df = pd.concat([df_2009,df_2010],ignore_index=True)
df.head()
```

```python
df=df[(df['Quantity']>0) & (df['Price']>0)]
```

```python
df.loc[:,'Total_price']=df['Quantity']*df['Price']
```

```python
df.loc[:,'Customer ID'] =df['Customer ID'].astype(str)
```

```python
df.info()
```

```python
df =df[~df['Invoice'].astype(str).str.startswith('C')]
```

```python
customer_df = df.groupby('Customer ID').agg({
    "Invoice": 'nunique',
    "Quantity":"sum",
    "Total_price":"sum",
    "InvoiceDate":['min','max']
})
```

```python
customer_df.columns =['Total_Orders','Total_Quantity','Total_Sales','First_Purchase_Date','Last_Purchase_Date']
```

```python
customer_df.reset_index()
```

```python
customer_df.head(10)
```

```python
customer_df['Life_Span'] =(customer_df['Last_Purchase_Date']-customer_df['First_Purchase_Date']).dt.days
```

```python
customer_df.reset_index()
```

```python
customer_df['Average_Order_Value']=(customer_df['Total_Sales']/customer_df['Total_Orders'])
```

```python
customer_df.reset_index()
```

```python
latest_Date = df['InvoiceDate'].max()

customer_df['Recency'] = (latest_Date -customer_df['Last_Purchase_Date']).dt.days
```

```python
customer_df = customer_df.reset_index()
```

```python
customer_df.head()
```

```python
customer_df['R_Score'] = pd.qcut(customer_df['Recency'],q=4,labels=[4,3,2,1]).astype(int)
```

```python
customer_df['F_Score'] = pd.qcut(customer_df['Total_Orders'].rank(method='first'),q=4,labels=[1,2,3,4]).astype(int)
```

```python
customer_df['M_Score'] = pd.qcut(customer_df['Total_Sales'],q=4,labels=[1,2,3,4]).astype(int)
```

```python
customer_df['RFM_Score'] = customer_df['R_Score'].astype(str) + customer_df['F_Score'].astype(str) + customer_df['M_Score'].astype(str)
```

```python
customer_df.head()
```

```python
def customer_segment(row):
    if row['RFM_Score'] == 444:
        return'Top Customers'
    elif row['R_Score'] >=3 and row['F_Score']>=3:
        return 'Loyal Customers'
    elif row['R_Score'] >=3 and row['M_Score'] >=3:
        return'Big Spenders'
    elif row['R_Score'] == 4:
        return 'Recent Customer'
    elif row['F_Score'] ==4:
        return 'Frequent Customer'
    elif row['M_Score'] ==4:
        return 'High Value'
    else:
        return 'Others'
```

```python
customer_df['Segment'] = customer_df.apply(customer_segment,axis=1)
```

```python
df['Cohort_Months']=df.groupby('Customer ID')['InvoiceDate'].transform('min').dt.to_period('M')
```

```python
df['Invoice_Months'] = df['InvoiceDate'].dt.to_period('M')
```

```python
df['Cohort_Index'] = (df['Invoice_Months'].dt.year - df['Cohort_Months'].dt.year) * 12 + (df['Invoice_Months'].dt.month - df['Cohort_Months'].dt.month) + 1
```

```python
df.head()
```

```python
cohort_data = df.groupby(['Cohort_Months','Cohort_Index'])['Customer ID'].nunique().reset_index()
```

```python
cohort_data.head()
```

```python
cohort_data.rename(columns={'Customer ID':'Total_Customers'},inplace=True)
```

```python
cohort_data.head()
```

```python
cohort_pivot = cohort_data.pivot(index='Cohort_Months',columns='Cohort_Index',values='Total_Customers')
```

```python
cohort_pivot.head()
```

```python
cohort_size =cohort_pivot.iloc[:,0]

retention_matrix = cohort_pivot.divide(cohort_size, axis=0).round(3)
```

```python
retention_matrix.head()
```

```python
import seaborn as sns

plt.figure(figsize=(15,6))
sns.heatmap(retention_matrix,annot=True,fmt='.0%',cmap='viridis')
plt.title('Cohort Retention Rate')
plt.ylabel("Cohort (First Purchase Date)")
plt.xlabel('Month Since First Purchase')
plt.show()
```


🔍 **What We Observed:**

Customer retention is good in the first month, but drops sharply after month 2 or 3. Some cohorts and segments stay active longer.

💡 **What Could Be Done:**

* Setup follow-up emails in month 2 and 3 to extend customer lifespan.
* Study why some cohorts retained better and replicate those practices.

```python
a_segment = (
       df.groupby('Segment')['Total_price'].sum()/
       df.groupby('Segment')['Invoice'].nunique()
).reset_index(name='Avg_Order_Value')
a_segment['Avg_Order_Value'] = a_segment['Avg_Order_Value'].round(3)
a_segment 
```

```python
purchase_freq = (
         df.groupby('Segment')['Invoice'].nunique()/
         df['Segment'].value_counts()
)         
purchase_freq = purchase_freq.reset_index(name= 'PurchaseFreq')
purchase_freq['PurchaseFreq'] = purchase_freq['PurchaseFreq'].round(6)
purchase_freq
 
```

```python
customer_Lifespan = df.groupby('Customer ID')['InvoiceDate'].agg(['min','max']).reset_index()
customer_Lifespan.columns = ['Customer ID','FirstPurchaseDate','LastPurchaseDate']
customer_Lifespan
```

```python
import numpy as np
customer_Lifespan['LifeSpan'] =(customer_Lifespan['LastPurchaseDate'] -customer_Lifespan['FirstPurchaseDate'])/np.timedelta64(1,'D')/30.44

customer_Lifespan['LifeSpan'] = customer_Lifespan['LifeSpan'].round(1)
customer_Lifespan
```

```python
# 1 Convert InvoiceData to DateTime

df['InvoiceData'] = pd.to_datetime(df['InvoiceDate'])

# 2 calculate First and Last purchase Date per customer
customer_Lifespan = df.groupby('Customer ID')['InvoiceDate'].agg(['min','max']).reset_index()
customer_Lifespan.columns=['Customer ID','FirstPurchaseDate','LastPurchaseDate']

# 3 Compute Lifespan in Months

customer_Lifespan['LifeSpan'] = (customer_Lifespan['LastPurchaseDate']-customer_Lifespan['FirstPurchaseDate']).dt.days//30

# 4 Merge it into main df
df = df.merge(customer_Lifespan[['Customer ID','LifeSpan']],on='Customer ID',how='left')
df


```

```python
avg_lifespan = df.groupby('Segment')['LifeSpan'].mean().reset_index(name='Avg_LifeSpan')
avg_lifespan['Avg_LifeSpan'] = avg_lifespan['Avg_LifeSpan'].round(1)
avg_lifespan
```

```python
ltv_df = a_segment.merge(purchase_freq, on='Segment').merge(avg_lifespan,on='Segment')
ltv_df
```

```python
ltv_df['LTV'] = ltv_df['PurchaseFreq'] * ltv_df['Avg_Order_Value'] * ltv_df['Avg_LifeSpan']
ltv_df
```

```python
plt.figure(figsize=(14,6))

xy =sns.barplot(data= ltv_df,x='Segment',y='Avg_Order_Value',hue='Segment',palette='viridis',legend=False)

for container in xy.containers:
    xy.bar_label(container,fmt='%.2f',label_type='edge',fontweight = 'bold')

plt.title('Average Order Value By Segment',fontweight='bold')
plt.xlabel('Segment',weight ='bold')
plt.ylabel('Average Order Value',weight ='bold') 
plt.xticks(rotation =0)
plt.tight_layout()
plt.show()
```

```python
# Group By Products And Month
monthly_qty = df.groupby(['Year_Month', 'Description'])['Quantity'].sum().reset_index()

# Select top 10 declining products: find those with highest drop
pivot_qty = monthly_qty.pivot(index='Year_Month',columns='Description',values='Quantity').fillna(0)

# Calculate Monthly Difference
Monthly_Diff = pivot_qty.diff().mean().sort_values().head(5)
Declining_products = Monthly_Diff.index

# Filter and plot

Declining_df=monthly_qty[monthly_qty['Description'].isin(Declining_products)].copy()
Declining_df['Year_Month'] = Declining_df['Year_Month'].astype(str)
Declining_df

plt.figure(figsize=(14,6))
sns.lineplot(data=Declining_df,x='Year_Month',y='Quantity',hue='Description',marker='o')
plt.title('Top 5 Declinig Products Over Time')
plt.xlabel('Month')
plt.ylabel('Quantity Sold')
plt.xticks(rotation = 45)
plt.show()
```


🔍 **What We Observed:**

Some products have been declining steadily over time — they may be losing interest or becoming outdated.

💡 **What Could Be Done:**

* Consider promotions or discounts for such items.
* If they continue to drop, they could be phased out or refreshed.

```python
df['Price_Bin'] = pd.cut(df['Price'], bins=[0, 2, 5, 10, 20, 50, 100, df['Price'].max()],
                         labels=['<2', '2-5', '5-10', '10-20', '20-50', '50-100', '100+'])

# Count purchases per bin and segment
price_sensitivity = df.groupby(['Segment', 'Price_Bin'],observed=False)['Quantity'].sum().unstack().fillna(0)
price_sensitivity

```

```python
# Heatmap of price sensitivity
plt.figure(figsize=(10, 6))
sns.heatmap(price_sensitivity, annot=True, fmt='.0f', cmap='magma')
plt.title('Price Sensitivity by Segment')
plt.xlabel('Price Range (£)')
plt.ylabel('Customer Segment')
plt.tight_layout()
plt.show()
```

🔍 **What We Observed:**

Most customers prefer cheaper products. But high-spending segments still buy expensive ones.

💡 **What Could Be Done:**

* Offer discounts to price-sensitive segments.
* Position premium items towards Top Customers who don’t mind paying more.
